```python
import pandas as pd

df = pd.read_csv('C:/Users/i5/Desktop/DATA ANALYST BOOTCAMP/data cleaning projects/UCI_Credit_Card.csv')
df.head()
```

```python
df.rename(columns={
    'PAY_0': 'PAY_1',
    'default.payment.next.month': 'default_next_month'
}, inplace=True)
df.head()
```

```python
df['SEX'] = df['SEX'].replace({1: 'Male', 2: 'Female'}).astype('category')

education_map = {1: 'Graduate School', 2: 'University', 3: 'High School', 4: 'Others', 5: 'Unknown', 6: 'Unknown', 0: 'Unknown'}
df['EDUCATION'] = df['EDUCATION'].replace(education_map).astype('category')

marriage_map = {1: 'Married', 2: 'Single', 3: 'Others', 0: 'Unknown'}
df['MARRIAGE'] = df['MARRIAGE'].replace(marriage_map).astype('category')

df['default_next_month'] = df['default_next_month'].replace({0: 'No Default', 1: 'Default'}).astype('category')

df
```

```python
pay_status_map = {
    -2: 'No Consumption',
    -1: 'Paid Duly',
     0: 'Paid Duly',
     1: '1 Month Delay',
     2: '2 Months Delay',
     3: '3 Months Delay',
     4: '4 Months Delay',
     5: '5 Months Delay',
     6: '6 Months Delay',
     7: '7 Months Delay',
     8: '8 Months Delay',
     9: '9+ Months Delay'
}
df['PAY_1'] = df['PAY_1'].replace(pay_status_map).astype('category')

pay_columns = ['PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

for col in pay_columns:
    df[col] = df[col].replace(pay_status_map).astype('category')

df
```

```python
duplicates = df.duplicated().sum()
print(f"Duplicates found: {duplicates}")

df = df.drop_duplicates()
```
