```python
df['name'] = df['name'].fillna("Unknown")
df['host_name'] = df['host_name'].fillna("Unknown")
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

df['last_review'] = pd.to_datetime(df['last_review'])
df['last_review'] = df['last_review'].fillna(pd.to_datetime("2000-01-01"))

df
```

```python
# Show only rows with invalid/non-numeric prices
invalid_prices = df[~df['price'].astype(str).str.replace(r'[\d.]', '', regex=True).str.strip().eq('')]
print(invalid_prices[['price']])
```

```python
result = df[df['price'].astype(str).str.contains(r'[^0-9]', regex=True)]

if result.empty:
    print("No invalid price values found (only digits).")
else:
    print(result)
```
